# NETFLIX

# Overview
The netflix.csv document contains a dataset of content available on Netflix, including both movies and TV shows. The dataset includes various attributes for each title, such as a unique identifier (show_id), the type of content (type), the title (title), and the director (director). It also lists the main cast members (cast), the country where the content was produced (country), the date it was added to Netflix (date_added), and the year of release (release_year). Additionally, the dataset provides the content rating (rating), the duration (in minutes for movies or number of seasons for TV shows), and the genres or categories the content belongs to (listed_in). The description column offers a brief overview of the content.

# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data

```python
data=pd.read_csv("netflix.csv")
data
```

# About Data

#### Column Descriptions:
        show_id: unique id given for shows
        type: the type of media
        title: the title of the media
        director: director of the media
        cast: actors/actresses involved in the show
        country: country where released
        date_added: date of release
        release_year: year of release
        rating: rating of the show
        duration: total running time
        listed_in: the genre of the media
        description: a little description on the media

# Data Visualization

```python
data['type'].value_counts().plot(kind='bar',color='red',edgecolor='black');
plt.title('Media shows on netflix');
```

```python
data['country'].value_counts().head(10).plot(kind='barh',color='pink');
plt.title('No. of films released in each country');
```

* most of the shows and films fall under TV-MA and TV-14 category of ratings

```python
def convert (date_added):
    return pd.to_datetime(date_added)
    
data['date_added']=data['date_added'].apply(convert)
```

```python
data['date_added_year']=data['date_added'].dt.year
```

```python
sns.jointplot(x='date_added_year',y='release_year',data=data,kind='scatter');
plt.title('distribution and relationship of release year and date added to netflix')
```

* there is a clustering of data points around recent years, suggesting that most movies and shows are added to Netflix within few years of their original release

```python
movies_df=data[data['type']=='Movie']
```

```python
def split(x):
    return x.split(' ')[0]
```

```python
movies_df['duration']=movies_df['duration'].apply(split)
```

```python
movies_df['duration']=movies_df['duration'].astype('int64')
```

```python
movies_df.groupby('title')['duration'].max().sort_values(ascending=False).head(10).plot(kind='bar',color='white',edgecolor='red')
plt.title('Movies with longest running time');
```

* the movie 'the school of mischief' is the movie with longest duration and 'No longer kids' at the second position

```python
movies_df.groupby(['director','title'])['duration'].max().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.1f%%')
plt.title('directors of long duration movies');
```

* this graph shows the directors of the movies with longest duration, Houssam El-DIn Mustafa's the school of mischief is the longest followed by other 9 films and their directors
